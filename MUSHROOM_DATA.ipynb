```python
import pandas as pd

df=pd.read_csv('/content/mushrooms.csv')
df
```

```python
x=df.drop(['class'],axis=1)
y=df['class']
```

```python
x.head()
```

```python
x.nunique()
```

```python
#Label encoding

from sklearn.preprocessing import LabelEncoder

le_x=LabelEncoder()

for i in x.columns:
  x[i]=le_x.fit_transform(x[i])
le_y=LabelEncoder()
y=le_y.fit_transform(y)

x.head()
```

```python
x=pd.get_dummies(x,columns=x.columns,drop_first=True)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

**Feature** **Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train=sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

x_train = pca.fit_transform(x_train)
x_test = pca.fit_transform(x_test)
```

**GaussianNB**

```python
from sklearn.naive_bayes import GaussianNB
gb=GaussianNB()
gb.fit(x_train,y_train)
```

```python
y_pred=gb.predict(x_test)
```

Making confusion matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score

cm=confusion_matrix(y_test,y_pred)
print('Accuracy score:', accuracy_score(y_test,y_pred))
cm
```

**HYPER TUNNING**

```python
import numpy as np
from sklearn.model_selection import GridSearchCV

parameters = {'var_smoothing': np.logspace(0,-9, num=100)}
Bayes = GridSearchCV(estimator=gb,
                     param_grid=parameters,
                     scoring='accuracy',
                     n_jobs=-1,
                     verbose=5,
                     cv=10)
Bayes.fit(x_train, y_train)
print('Train Accuracy : %.3f'%Bayes.best_estimator_.score(x_train, y_train))
print('Test Accuracy : %.3f'%Bayes.best_estimator_.score(x_test, y_test))
print('Best estimator : ',Bayes.best_estimator_)
print('Best Accuracy score:',Bayes.best_score_)
predictions = Bayes.best_estimator_.predict(x_test)
print('Accuracy score:', accuracy_score(y_test,predictions))
cm=confusion_matrix(y_test,predictions)
cm
```

**FITTING BERNOULLINB**

```python
from sklearn.naive_bayes import BernoulliNB
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
y_pred=bnb.predict(x_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
print('Accuracy score:', accuracy_score(y_test,y_pred))
cm
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV

parameters = {'alpha':[0.01, 0.1, 0.5, 1.0, 10.0]}
bnb_param = GridSearchCV(estimator=bnb,
                     param_grid=parameters,
                     scoring='accuracy',
                     n_jobs=-1,
                     verbose=5,
                     cv=10)
bnb_param.fit(x_train, y_train)
print('Train Accuracy : %.3f'%bnb_param.best_estimator_.score(x_train, y_train))
print('Test Accuracy : %.3f'%bnb_param.best_estimator_.score(x_test, y_test))
print(bnb_param.best_estimator_)
print('Best Accuracy score:',bnb_param.best_score_)
predictions = bnb_param.best_estimator_.predict(x_test)
print('Accuracy score:', accuracy_score(y_test,predictions))
cm=confusion_matrix(y_test,predictions)
cm
```

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Instantiate dt
dt = DecisionTreeClassifier(min_samples_leaf=0.13, random_state=42)

# Instantiate LogisticRegression
lr = LogisticRegression(random_state=42)

# Instantiate random forest
rf = RandomForestClassifier( random_state = 42)

# Instantiate SVM
from sklearn.svm import SVC
svc = SVC(kernel='rbf',random_state=42)

# Instantiate KNN
from sklearn.neighbors import KNeighborsClassifier as KNN
knn = KNN(n_neighbors=7)
```

```python
# Define the list classifiers
classifiers = [ ('K Nearest Neighbours', knn), ('Classification Tree', dt), ('BernoulliNB',bnb_param),('GaussianNB',Bayes),('Random Forest',rf),('SVC',svc),('Logistic regresion',lr)]

from sklearn.metrics import accuracy_score,roc_auc_score
for clf_name,clf in classifiers:
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)

    accuracy=accuracy_score(y_test,y_pred)

    print('{:s} : {:.4f}'.format(clf_name, accuracy))
```

```python

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

knn = KNN()
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print('Accuracy score:', accuracy_score(y_test,y_pred))
parameters={'metric':['minkowski','euclidean','manhattan'],
            'n_neighbors':[1,2,3,4,5,6,8,9,10,11,12,13,14,15],
            'weights':['uniform','distance']}
knn_param = GridSearchCV(estimator=knn,
                     param_grid=parameters,
                     scoring='accuracy',
                     n_jobs=-1,
                     verbose=5,
                     cv=10)
knn_param.fit(x_train, y_train)
predictions = knn_param.best_estimator_.predict(x_test)
print('Best parameters:',knn_param.best_params_)
print(knn_param.best_estimator_)
print('Train Accuracy : %.3f'%knn_param.best_estimator_.score(x_train, y_train))
print('Test Accuracy : %.3f'%knn_param.best_estimator_.score(x_test, y_test))
print('Best Accuracy score:',knn_param.best_score_)
print('Accuracy score:', accuracy_score(y_test,predictions))

print(confusion_matrix(y_test,predictions ))

from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
